# **ðŸ§¹ 3. Data Cleaning**

## Mapping and Convert Columns with Range data to Mid point of Ranges

Simplifies data by standardizing names and converting range-based columns to midpoints. This enhances consistency and enables more straightforward analysis.

```python
department_mapping = {
    'Engineering - CS / CSE / CSC / Similar to CS': 'CS/IT Engineering',
    'Other': 'Other',
    'Engineering - EEE/ ECE / Similar to EEE': 'EEE/ECE Engineering',
    'Environmental and Life Sciences': 'Env/Life Sciences',
    'Business and Entrepreneurship Studies': 'Business/Entrepreneurship',
    'Biological Sciences': 'Biological Sciences',
    'Engineering - Civil Engineering / Similar to CE': 'Civil Engineering',
    'Engineering - Mechanical Engineering / Similar to ME': 'Mechanical Engineering'
}

Academic_year_mapping ={
    'Fourth Year or Equivalent': 4,
    'First Year or Equivalent': 1,
    'Third Year or Equivalent': 3,
    'Second Year or Equivalent': 2,
    'Other': 5           # Assume fifth year
}

age_mapping = {
    '18-22': 20,
    '23-26': 25,
    '27-30': 28,
    'Below 18': 18, # Cap at 18 as records are minimal
    'Above 30': 30 # Cap at 30 as records are minimal
}

cgpa_mapping = {
    'Below 2.50': 2.0,
    '2.50 - 2.99': 2.75,
    '3.00 - 3.39': 3.20,
    '3.40 - 3.79': 3.60,
    '3.80 - 4.00': 3.90,
    'Other': None           # Replace with None to be handled later
}

df.loc[:, 'Department'] = df['Department'].map(department_mapping)
df.loc[:, 'Academic Year'] = df['Academic Year'].map(Academic_year_mapping)
df.loc[:, 'Age'] = df['Age'].map(age_mapping)
df.loc[:, 'Current CGPA'] = df['Current CGPA'].map(cgpa_mapping)
```

## Handle Records with Limited Information
1. Gender: Replace all instances of 'Prefer not to say' in the Gender column with mode.
2. Current CGPA: Replace Current CGPA with a representative average
3. Department: Replace missing values with the most common department


```python
# Handle records with limited information in gender column
df['Gender'] = df['Gender'].replace('Prefer not to say', df['Gender'].mode()[0])
```

```python
# Impute missing values in 'Current CGPA' with the mean
mean_cgpa = df['Current CGPA'].mean()
df['Current CGPA'] = df['Current CGPA'].fillna(mean_cgpa)
```

```python
# Impute missing values in 'Department' with mode
most_common_department = df['Department'].mode()[0]
df['Department'] = df['Department'].fillna(most_common_department)
```

## Analyze the distribution of key demographic and academic characteristics among the student population.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the columns to visualize
cols_to_visualize = ['Gender', 'Department', 'Academic Year', 'Waiver/Scholarship']

# Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, col in enumerate(cols_to_visualize):
    ax = axes[i]
    sns.countplot(data=df, x=col, hue=col, palette='viridis', ax=ax, legend=False)

    total = len(df[col])
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}\n({height/total:.1%})',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9, color='black')

    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = ['Depression Label', 'Anxiety Label', 'Stress Label']

labels_order = {
    "Depression Label": ["No Depression", "Minimal Depression", "Mild Depression", "Moderate Depression", "Moderately Severe Depression", "Severe Depression"],
    "Anxiety Label": ["Minimal Anxiety", "Mild Anxiety", "Moderate Anxiety", "Severe Anxiety"],
    "Stress Label": ["Low Stress", "Moderate Stress", "High Perceived Stress"]
}

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for i, label_col in enumerate(labels):
    ax = axes[i]
    sns.countplot(data=df, x=label_col, hue='Waiver/Scholarship', palette='viridis', ax=ax,
                  order=labels_order[label_col])
    ax.set_title(f'{label_col} by Waiver/Scholarship')
    ax.set_xlabel(label_col)
    if i == 0:
        ax.set_ylabel('Count')
    else:
        ax.set_ylabel('')
    ax.tick_params(axis='x', rotation=45)

    # Add counts on top of bars
    total = len(df)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
labels = ['Depression Label', 'Anxiety Label', 'Stress Label']

labels_order = {
    "Depression Label": ["No Depression", "Minimal Depression", "Mild Depression", "Moderate Depression", "Moderately Severe Depression", "Severe Depression"],
    "Anxiety Label": ["Minimal Anxiety", "Mild Anxiety", "Moderate Anxiety", "Severe Anxiety"],
    "Stress Label": ["Low Stress", "Moderate Stress", "High Perceived Stress"]
}

# Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for i, label_col in enumerate(labels):
    ax = axes[i]

    heatmap_data = df.groupby(['University', label_col]).size().unstack(fill_value=0)

    valid_order = [label for label in labels_order[label_col] if label in heatmap_data.columns]
    heatmap_data = heatmap_data.reindex(columns=valid_order)

    heatmap_percent = heatmap_data.div(heatmap_data.sum(axis=1), axis=0) * 100

    annot = heatmap_percent.round(1).astype(str) + '%'

    sns.heatmap(heatmap_percent, annot=annot, fmt='', cmap='viridis', linewidths=.5,
                ax=ax, cbar=(i == 2))

    ax.set_title(label_col.replace(' Label', ''))
    ax.set_xlabel('')
    if i == 0:
        ax.set_ylabel('University')
    else:
        ax.set_ylabel('')
    ax.tick_params(axis='x', rotation=45)
    ax.tick_params(axis='y', rotation=0)

plt.suptitle('Mental Health Label Distribution by University (Percentage)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Insights: Anxiety Levels by University and Scholarship Status

1. **General Observation**  
   Across all universities, **Moderate and Minimal Anxiety** are the most commonly reported categories, with relatively fewer students reporting Severe Anxiety.

2. **Scholarship Holders**  
   - Students **with scholarships** generally show **lower levels of Severe Anxiety** across universities.
   - Their anxiety distribution skews more toward **Minimal and Mild Anxiety**, suggesting a potential buffering effect of financial aid.
   - Institutions like **North South University (NSU)** and **Ahsanullah University of Science and Technology (AUST)** show particularly low severe anxiety among scholarship recipients.

3. **Non-Scholarship Students**  
   - Students **without scholarships** show a **higher proportion of Moderate and Severe Anxiety**.
   - **DUET**, **RUET**, and **IUB** have noticeable peaks in anxiety for this group, indicating stress may be more prevalent in these institutions among financially unsupported students.

4. **University-Level Differences**  
   - Some universities show a **wide spread across all anxiety levels** (e.g., IUB, DUET), while others cluster around minimal to mild levels.
   - This suggests that **institutional context**, such as academic pressure or support systems, may influence mental health differently.

> **Interpretation Tip**: Financial aid may act as a protective factor against anxiety, but the impact appears **university-dependent**, likely influenced by other variables such as curriculum pressure, academic culture, or student support infrastructure.

##Explore how gender, department, academic year, and age collectively influence suicidal thoughts.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Ensure Suicidal Thoughts is categorical
df['Suicidal Thoughts'] = df['Suicidal Thoughts'].astype(str)

# Function to add count on top of bars
def add_counts(ax):
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=9, padding=2)

# 1. Suicidal Thoughts by Gender
plot1 = sns.catplot(
    data=df, x="Suicidal Thoughts", kind="count", hue="Gender",
    height=4, aspect=1.2, palette="viridis"
)
plot1.set(
    title="Suicidal Thoughts Distribution by Gender",
    xlabel="Suicidal Thoughts", ylabel="Count"
)
add_counts(plot1.ax)

# 2. Suicidal Thoughts by Department and Gender
plot2 = sns.catplot(
    data=df, x="Suicidal Thoughts", kind="count", hue="Gender", col="Department",
    col_wrap=3, height=4, aspect=1.2, palette="viridis", sharey=False
)
plot2.set_titles("{col_name}").set_axis_labels("Suicidal Thoughts", "Count")
for ax in plot2.axes.flat:
    add_counts(ax)

# 3. Suicidal Thoughts by Academic Year and Gender
plot3 = sns.catplot(
    data=df, x="Suicidal Thoughts", kind="count", hue="Gender", col="Academic Year",
    col_wrap=3, height=4, aspect=1.2, palette="viridis", sharey=False
)
plot3.set_titles("Year: {col_name}").set_axis_labels("Suicidal Thoughts", "Count")
for ax in plot3.axes.flat:
    add_counts(ax)

# 4. Suicidal Thoughts by Age Group and Gender
age_mapping = {
    'Below 18': 18,
    '18-22': 20,
    '23-26': 25,
    '27-30': 28,
    'Above 30': 30
}
reverse_age_mapping = {v: k for k, v in age_mapping.items()}
age_order = ['Below 18', '18-22', '23-26', '27-30', 'Above 30']

# Create a temporary copy for plotting
df_plot = df.copy()
df_plot['Age Group'] = df_plot['Age'].map(reverse_age_mapping)
df_plot['Age Group'] = pd.Categorical(df_plot['Age Group'], categories=age_order, ordered=True)

plot4 = sns.catplot(
    data=df_plot,
    x="Suicidal Thoughts",
    kind="count",
    hue="Gender",
    col="Age Group",
    col_wrap=3,
    height=4,
    aspect=1.2,
    palette="viridis",
    sharey=False
)
plot4.set_titles("Age Group: {col_name}").set_axis_labels("Suicidal Thoughts", "Count")
for ax in plot4.axes.flat:
    add_counts(ax)

plt.show()
```

Insights: Suicidal Thoughts by Demographic and Academic Variables

- **Gender Differences**  
  Male students consistently report **higher levels of suicidal thoughts** than female students, across all severity levels. The gap is especially evident in moderate to high ranges.

- **Departmental Variations**  
  Suicidal thoughts appear more frequently in certain departments, especially **CS/IT** and **EEE**. These technical programs show a higher concentration of mild to moderate suicidal ideation, particularly among male students.

- **Academic Year Impact**  
  The prevalence of suicidal thoughts **fluctuates across academic years**, with noticeable peaks in **Year 1** and **Year 3**. This may reflect transitional challenges during the early phase and academic pressure during final project-heavy years.

- **Age-Related Patterns**  
  The distribution of suicidal ideation **varies by age group**, with younger students (ages 20â€“22) generally exhibiting higher reported frequencies, possibly due to adjustment stress, identity development, or lack of support.

- **High-Risk Subgroups Identified**  
  - **Male Engineering Students**: Especially those in **CSE/IT** and **ECE** departments, consistently report higher frequencies of suicidal thoughts than female counterparts.
  - **First-Year Students**: Suicidal thoughts are notably present in **Year 1**, potentially due to academic adjustment, social isolation, or transition stress from high school to university.

>**Interpretation Tip**: The data reveals that suicidal ideation is influenced by a complex interplay of gender, department, academic standing, and age â€” suggesting that **targeted mental health interventions** should focus on high-risk clusters like **male CS/Engineering students** and **first-year cohorts**.

##Assess the impact of age, academic year, and current CGPA on depression, anxiety, and stress.

```python
import matplotlib.pyplot as plt
import seaborn as sns

palette = "viridis"
label_cols = ['Depression Label', 'Anxiety Label', 'Stress Label']
hue_cols = ['Age', 'Academic Year', 'Current CGPA']

fig, axes = plt.subplots(len(label_cols), len(hue_cols), figsize=(18, 15), constrained_layout=True)

for i, label_col in enumerate(label_cols):
    for j, hue_col in enumerate(hue_cols):
        ax = axes[i, j]
        sns.countplot(
            data=df,
            x=label_col,
            hue=hue_col,
            palette=palette,
            dodge=True,
            ax=ax
        )
        ax.set_title(f"{label_col} by {hue_col}")
        ax.set_xlabel(label_col)
        ax.set_ylabel("Count")
        ax.tick_params(axis='x', rotation=45)

        # Add count labels on top of each bar
        for container in ax.containers:
            ax.bar_label(container, label_type='edge', fontsize=8, padding=2)

        # Set legend inside plot area
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles, labels, title=hue_col, loc='best', fontsize='small', title_fontsize='medium')

plt.show()
```

```python
categorical_columns = df.select_dtypes(include="object").columns
categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder

encoder_dict = {}
mapping_dict = {}

labels = ['Depression Label', 'Anxiety Label', 'Stress Label']
labels_order = {
    "Depression Label": ["No Depression", "Minimal Depression", "Mild Depression", "Moderate Depression", "Moderately Severe Depression", "Severe Depression"],
    "Anxiety Label": ["Minimal Anxiety", "Mild Anxiety", "Moderate Anxiety", "Severe Anxiety"],
    "Stress Label": ["Low Stress", "Moderate Stress", "High Perceived Stress"]
}

for label_col in labels:
    df[label_col] = pd.Categorical(df[label_col], categories=labels_order[label_col], ordered=True)

for col in categorical_cols:
    # Map the labels in sequence
    if col in labels:
        mapping = {cat: code for code, cat in enumerate(df[col].cat.categories)}
        mapping_dict[col] = mapping
        df[col] = df[col].cat.codes
        encoder_dict[col] = None
    else:
        # For other categorical columns, use LabelEncoder
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
        encoder_dict[col] = encoder
        mapping_dict[col] = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

for col, mapping in mapping_dict.items():
    print(f"\n{col} Encoding Mapping:")
    for label, code in mapping.items():
        print(f"{code}. {label}")
```

## Select Best Model To Predict Anxiety, Stress and Depression

To systematically evaluate and compare machine learning models for predicting student mental health conditions, we developed a unified evaluation pipeline that:

- Iterates across three classification tasks: **Anxiety**, **Stress**, and **Depression**
- Applies three selected models with fine-tuned parameters:
  - Logistic Regression
  - Random Forest
  - Support Vector Machine (SVM)

### Evaluation Metrics
For each model-task combination, we compute the following metrics:
- **Accuracy**
- **Macro F1-score** (primary metric for imbalanced multi-class classification)
- **Precision & Recall**
- **Log Loss**
- **Cross-Validated F1 (Mean Â± Std)**

### Overfitting Check
We include a three-level overfitting flag based on both:
- The gap between **Train F1** and **Test F1**
- The gap between **Test F1** and **Cross-Validated F1 Mean**

An ensemble gap score is computed:

**gap_score = |Train F1 âˆ’ Test F1| + |Test F1 âˆ’ CV F1|**

Thresholds:
- âŒ **Severe**: `gap_score > 0.3` or `Train-Test Gap > 0.2`
- âš ï¸ **Likely**: `gap_score > 0.15` or `Train-Test Gap > 0.1`
- âœ… **No**: otherwise

### Setup Highlights
- Models are trained using an **oversampled training set**
- Evaluation is consistently performed on a **shared test set**
- A **confusion matrix** is plotted for each model to visualize misclassifications
- **Label simplification** (e.g., 5-class Depression) is optionally applied to reduce label noise and improve generalization

This setup ensures fair model comparison under the same experimental conditions, helping us identify the most balanced and high-performing classifier for each mental health classification task.

```python
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, log_loss
)
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ========== Define Dataset ==========
conditions = {
    "Anxiety": (X_train_Anxiety, y_train_Anxiety, y_test_Anxiety, X_test_Anxiety),
    "Stress": (X_train_Stress, y_train_Stress, y_test_Stress, X_test_Stress),
    "Depression": (X_train_Depression, y_train_Depression, y_test_Depression, X_test_Depression)
    # "Depression": (X_train_Depression, y_train_Depression_3class, y_test_Depression_3class, X_test_Depression)
}

# ========== Define Models ==========

models = {
    "Logistic Regression": LogisticRegression(
        max_iter=500,
        class_weight='balanced',
        multi_class='multinomial',
        solver='lbfgs',
        random_state=42
    ),

    # "XGBoost": XGBClassifier(
    #     objective='multi:softmax', num_class=3, eval_metric='mlogloss', use_label_encoder=False,
    #     learning_rate=0.05, n_estimators=100, max_depth=3,
    #     min_child_weight=5, gamma=3, subsample=0.7, colsample_bytree=0.7, random_state=42),

    # "CatBoost": CatBoostClassifier(
    #     loss_function="MultiClass", depth=6, l2_leaf_reg=4.0, iterations=300, random_seed=42, verbose=0),

    # "MLP": MLPClassifier(
    #     hidden_layer_sizes=(30,), alpha=0.05, early_stopping=True, n_iter_no_change=5, max_iter=120, random_state=42),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=4,
        max_features='log2',
        class_weight='balanced',
        random_state=42
    ),
    "SVM": SVC(
        kernel='rbf',
        C=1.0,
        gamma='scale',
        class_weight='balanced',
        probability=True,
        random_state=42
    ),
}

# ========== Evaluation ==========
results = []

for condition, (X_train, y_train, y_test, X_test_sel) in conditions.items():
    print(f"\nðŸ“˜ Condition: {condition}")

    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test_sel)
        y_train_pred = model.predict(X_train)

        # Try log-loss only if model has predict_proba
        try:
            y_prob = model.predict_proba(X_test_sel)
            loss = log_loss(y_test, y_prob)
        except:
            loss = np.nan

        # Evaluation Metrics
        acc = accuracy_score(y_test, y_pred)
        f1_macro = f1_score(y_test, y_pred, average="macro")
        prec_macro = precision_score(y_test, y_pred, average="macro")
        recall_macro = recall_score(y_test, y_pred, average="macro")
        cm = confusion_matrix(y_test, y_pred)

        f1_train = f1_score(y_train, y_train_pred, average="macro")

        # Cross-validation
        cv_score = cross_val_score(model, X_train, y_train, cv=5, scoring='f1_macro')
        cv_mean = np.mean(cv_score)
        cv_std = np.std(cv_score)

        # Overfitting check
        train_test_gap = abs(f1_train - f1_macro)
        cv_gap = abs(f1_macro - cv_mean)
        gap_score = train_test_gap + cv_gap  # Ensemble metric

        if gap_score > 0.3 or train_test_gap > 0.2:
            overfit_flag = "âŒ Severe"
        elif gap_score > 0.15 or train_test_gap > 0.1:
            overfit_flag = "âš ï¸ Likely"
        else:
            overfit_flag = "âœ… No"

        # Store results
        results.append({
            "Task": condition,
            "Model": name,
            "Test F1": round(f1_macro, 4),
            "Accuracy": round(acc, 4),
            "Recall": round(recall_macro, 4),
            "Overfitting?": overfit_flag
        })

        # Confusion Matrix Plot
        print(f"\nðŸ” {name} on {condition}")
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
        plt.title(f"{condition} - {name} Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.tight_layout()
        plt.show()

        # Classification Report
        print("\nðŸ“Š Classification Report:")
        print(classification_report(y_test, y_pred, digits=3))

        # Summary
        print(f"Accuracy: {acc:.4f} | F1: {f1_macro:.4f} | Precision: {prec_macro:.4f} | Recall: {recall_macro:.4f}")
        print(f"Log Loss: {loss if not np.isnan(loss) else 'N/A'} | CV F1: {cv_mean:.4f} Â± {cv_std:.4f}")
        print(f"Overfitting? â†’ {overfit_flag}")
```

##Final Model Training & Evaluation Workflow

This section outlines the finalized and optimized pipeline for training, predicting, and evaluating logistic regression models on three mental health conditions: Anxiety, Stress, and Depression. After applying feature selection and oversampling to improve generalization and reduce class imbalance, Logistic Regression was selected as the final model for its high interpretability, strong performance, and low overfitting risk.

The workflow trains separate models using task-specific selected features and balanced training sets. It generates predictions on isolated test sets and evaluates the results using Accuracy, Macro F1-score, Precision, Recall, and Confusion Matrices. This structured approach ensures reproducible, interpretable, and deployment-ready models suitable for mental health prediction applications.


```python
# Final Model Training & Evaluation Workflow (Post Feature Selection & Oversampling Version)

# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    recall_score, classification_report, confusion_matrix
)
import warnings
warnings.filterwarnings('ignore')

# Define Final Datasets
conditions = {
    "Anxiety": (X_train_Anxiety, y_train_Anxiety, X_test_Anxiety, y_test_Anxiety),
    "Stress": (X_train_Stress, y_train_Stress, X_test_Stress, y_test_Stress),
    "Depression": (X_train_Depression, y_train_Depression, y_test_Depression, X_test_Depression)
    # "Depression": (X_train_Depression, y_train_Depression_3class, X_test_Depression, y_test_Depression_3class)
}

# Initialize and Train Final Logistic Regression Models
final_models = {}
for label, (X_train, y_train, _, _) in conditions.items():
    model = LogisticRegression(
        max_iter=500, class_weight='balanced',
        multi_class='multinomial', solver='lbfgs', random_state=42
    )
    model.fit(X_train, y_train)
    final_models[label] = model

# Unified Prediction
def predict_all_labels(X_dict):
    return {
        "Anxiety Label": final_models["Anxiety"].predict(X_dict["Anxiety"]).ravel(),
        "Stress Label": final_models["Stress"].predict(X_dict["Stress"]).ravel(),
        "Depression Label": final_models["Depression"].predict(X_dict["Depression"]).ravel()
    }

# Predict on Individual Test Sets
X_test_dict = {
    "Anxiety": X_test_Anxiety,
    "Stress": X_test_Stress,
    "Depression": X_test_Depression
}
predictions = predict_all_labels(X_test_dict)

# Format as DataFrame
pred_df = pd.DataFrame(predictions)
y_test_df = pd.DataFrame({
    "Anxiety Label": y_test_Anxiety,
    "Stress Label": y_test_Stress,
    # "Depression Label": y_test_Depression_3class
    "Depression Label": y_test_Depression
})

# Evaluation (Confusion Matrix & Classification Report)
for label in ["Anxiety Label", "Stress Label", "Depression Label"]:
    y_true = y_test_df[label]
    y_pred = pred_df[label]

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix â€“ {label}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    print(f"\nðŸ“Š Classification Report for {label}:")
    print(classification_report(y_true, y_pred, digits=3))

def plot_feature_importance_subplot(ax, model, feature_names, title):
    if hasattr(model, "coef_"):
        importances = np.abs(model.coef_).mean(axis=0)
        indices = np.argsort(importances)[::-1]

        sorted_features = np.array(feature_names)[indices][:20]
        sorted_importances = importances[indices][:20]

        sns.barplot(
            x=sorted_importances,
            y=sorted_features,
            palette="viridis",
            ax=ax
        )

        ax.set_title(title, fontsize=14, weight='bold')
        ax.set_xlabel("Feature Importance Score", fontsize=12)
        ax.set_ylabel("Feature", fontsize=12)
        ax.tick_params(axis='x', labelsize=10)
        ax.tick_params(axis='y', labelsize=9)

        # Add value labels
        for i, value in enumerate(sorted_importances):
            ax.text(value + 0.002, i, f'{value:.3f}', color='black', va='center', fontsize=8)

# === Create subplots (3-in-1 row) ===
fig, axes = plt.subplots(1, 3, figsize=(24, 10), sharex=False)

plot_feature_importance_subplot(axes[0], final_models["Anxiety"], anxiety_features, "Feature Importance â€“ Anxiety")
plot_feature_importance_subplot(axes[1], final_models["Stress"], stress_features, "Feature Importance â€“ Stress")
plot_feature_importance_subplot(axes[2], final_models["Depression"], depression_features, "Feature Importance â€“ Depression")

plt.tight_layout()
plt.show()
```
